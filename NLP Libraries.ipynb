```python
from transformers import pipeline
```

```python
nlp=pipeline('sentiment-analysis')
nlp
```

```python
nlp('hii,how are you')
```

```python
 nlp('i am')
```

```python
nlp1=pipeline('question-answering')
nlp1
```

```python
nlp1(context='my age is 24',question='What is your age')
```

```python
nlp1(context='my name is athul',question='what is your name')
```

```python
nlp3=pipeline('ner')
nlp3
```

```python
nlp3('My name is athul')
```
